```python
#ì›¹ìº 
import cv2

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("cant open webcam")
    exit()

print("start, if you waunt quit, press q")

while True:
    ret, frame = cap.read()
    if not ret:
        print("cant take frame. exit")
        break

    results = model.predict(frame)
    result = results[0].boxes

    for boxes in result:
        
        #x1, y1, x2, y2, confidence, class_id = result
        xyxy = np.array(boxes.xyxy[0]).astype(int)
        cls = boxes.cls.item()
        conf = boxes.conf.item()
        if conf > 0.5:
            label = f"{cls}: {conf:.2f}"
            color = (0, 255, 0)
            cv2.rectangle(frame, (xyxy[0], xyxy[1]), (xyxy[2], xyxy[3]), color, 2)
            cv2.putText(frame, label, (xyxy[0], xyxy[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    cv2.imshow("real-time inference", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
