```python
#archive에서 실행

import os
import shutil

folder = ['train', 'valid', 'test']

for f in folder:
    # 원본 폴더와 저장할 폴더 설정
    labels_src = f"{f}/labels"  # 원본 폴더 경로
    labels_dest = f"{f}/selected_labels"  # 저장할 폴더 경로
    images_src = f"{f}/images"
    images_dest = f"{f}/selected_images"
    
    # 저장할 폴더가 없으면 생성
    os.makedirs(labels_dest, exist_ok=True)
    os.makedirs(images_dest, exist_ok=True)
    
    # 파일 목록 가져오기
    file_list = [f for f in os.listdir(labels_src) if f.endswith('.txt')]
    
    # 선택된 파일 이름 (숫자 부분) 저장
    selected_files = []
    
    for file_name in file_list:
        file_path = os.path.join(labels_src, file_name)
        
        # 파일 내용 읽기
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        
        # 조건에 따라 선택
        if content and content[0] in ('0', '1', '5'):
            selected_files.append(file_name.split('.')[0])
    
    # 선택된 .txt와 동일한 이름의 .jpg 복사
    for file_idx in selected_files:
        txt_file = f"{file_idx}.txt"
        txt_src = os.path.join(labels_src, txt_file)
        txt_dest = os.path.join(labels_dest, txt_file)
        if os.path.exists(txt_src):
            shutil.copy(txt_src, txt_dest)
            print(f"{txt_file} 파일을 저장했습니다.")
        
        jpg_file = f"{file_idx}.jpg"
        jpg_src = os.path.join(images_src, jpg_file)
        jpg_dest = os.path.join(images_dest, jpg_file)
        if os.path.exists(jpg_src):
            shutil.copy(jpg_src, jpg_dest)
            print(f"{jpg_file} 파일을 저장했습니다.")
```
