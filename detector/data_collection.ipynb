```python
from PIL import Image
import os

folders = ['train', 'test', 'valid']

for f in folders:
    for t in os.listdir(f):
        if "image" not in t:
            continue
        t_path = os.path.join(f, t)
        for img in os.listdir(t_path):
            img_path = os.path.join(t_path, img)

            # 이미지 열기
            image = Image.open(img_path)

            # 다운스케일링할 크기 (예: 640 -> 4`6`)
            new_width = 160
            new_height = 160

            # 이미지 크기 조정
            resized_image = image.resize((new_width, new_height), Image.LANCZOS)

            # 결과 저장
            resized_image.save(img_path)

            print("Resized {}".format(img_path))
```
